```python
import json
import pandas as pd
import folium
```

```python
data = json.load(open('Foursquare.json'))
df = pd.DataFrame(data["results"])
```

```python
df
```

```python
df['geocodes'][1]
```

```python
df['latitude'] = df['geocodes'].apply(lambda x: x['main']['latitude'])
df['longitude'] = df['geocodes'].apply(lambda x: x['main']['longitude'])
```

```python
import folium 

center_lat = df['latitude'].mean()
center_lng = df['longitude'].mean()

restaurant_map = folium.Map(location=[center_lat, center_lng], zoom_start=18)

for idx, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],  
        popup=row['name']
    ).add_to(restaurant_map)

restaurant_map
```
